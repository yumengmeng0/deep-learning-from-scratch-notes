```python
X[0][1]  # (0, 1)的元素
```

## 1.6 Matplotlib

Matplotlib是用于绘制图形的库。

### 1.6.1 绘制简单图形

```python
import matplotlib.pyplot as plt

# 生成数据
x = np.arange(0, 6, 0.1)
y = np.sin(x)

# 绘制图形
plt.plot(x, y)
plt.show()
```

### 1.6.2 pyplot的功能

```python
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle='--', label='cos')
plt.xlabel('x')  # x轴标签
plt.ylabel('y')
plt.title('sin & cos')  # 标题
plt.legend() # 显示图例（左下角）
plt.show()
```

### 1.6.3 显示图像

```python
from matplotlib.image import imread

img = imread('../dataset/lena.png')
plt.imshow(img)
plt.show()
```
